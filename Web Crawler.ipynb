## BeautifulSoup

```
import urllib
import requests as _requests
from bs4 import BeautifulSoup as _BeautifulSoup
import os
import pandas as pd


df = pd.DataFrame(columns=['Links', 'Lyrics', 'Mood'])
file = open("test_data_demo.txt", 'w')

url = "https://lyrics.fandom.com/wiki/Category:Language/Spanish?from=Q"
response = _requests.get(url, timeout=None)
soup = _BeautifulSoup(response.content, "html.parser")
data = soup.findAll('li',attrs={'class':'category-page__member'})
count=0
for div in data:
    links = div.findAll('a')
    for a in links:
        lyric_link = "https://lyrics.fandom.com/"+a['href']
        count+=1
        response_lyrics = _requests.get(lyric_link, timeout=None)
        lyric_soup = _BeautifulSoup(response_lyrics.content, "html.parser")
        lyricboxes = lyric_soup.findAll('div',attrs={'class':"lyricbox"})
        
        for lyricbox in lyricboxes:
          for br in lyricbox.findAll('br'):
            br.replace_with('\n')
          for div in lyricbox.findAll('div'):
            div.replace_with("")

        for lyricbox in lyricboxes:
          df = df.append({'Links': lyric_link.strip(), 'Lyrics': lyricbox.text.strip()}, ignore_index=True)
      


file.write(df.to_csv(header = False, index = False))
file.close()
```
